```python
# 4. Write a Python program to calculate the sum of three given numbers. If the values are equal,
# return three times their sum.
a=int(input("Enter 1st number :"))
b=int(input("Enter 2st number :"))
c=int(input("Enter 3st number :"))
sum=a+b+c
if (a==b==c) :
    print(sum*3)
else :
    print(sum)
```

```python
# 5. Write a Python program to solve (x + y) * (x + y).
# Test Data : x = 4, y = 3
# Expected Output : (4 + 3) ^ 2) = 49
x=int(input("Enter x  :"))
y=int(input("Enter y  :"))
res=(x+y)*(x+y)
print('(',x,'+',y,') ^ 2 = ',res)
```

```python
# 6. Write a Python program to compute the future value of a specified principal amount, rate of
# interest, and number of years.
# Test Data : amt = 10000, int = 3.5, years = 7
# Expected Output : 12722.79
amt=float(input("amt  :"))
rate=float(input("ins  :"))
year=int(input("years :"))
res = amt * (1 + rate / 100) ** year
print(round(res,2))
```

```python
# 7. Write a Python program to parse a string to float or integer.
s = input("Enter a number: ")

if '.' in s:
    res= float(s)
    print("Float :", res)
else:
    res = int(s)
    print("Integer :", res)
```

```python
#9. Write a Python program to calculate sum of digits of a number.
n=int(input("Enter n value :"))
x=n
total=0
while(x!=0) :
    a=x%10
    total=total+a
    x=x//10
print(total)
```

```python
# 13. Write a program which will find all such numbers which are divisible by 7 but are not a multiple
# of 5, between 2000 and 3200 (both included).
a=2000
b=3200
for i in range(a,b+1) :
    if (i%7==0 and i%5!=0):
        print(i,end=" ") 
```

```python
# 14. Write a program that calculates and prints the value according to the given formula:
# Q = Square root of [(2 * C * D)/H]
# Following are the fixed values of C and H:
# C is 50. H is 30. D is the variable whose values should be input to your program in a comma-separated sequence.
# Example
# Let us assume the following comma separated input sequence is given to the program:
# 100,150,180
# The output of the program should be: 18,22,24
# Hints:
# If the output received is in decimal form, it should be rounded off to its nearest value (for
# example, if the output received is 26.0, it should be printed as 26)
# In case of input data being supplied to the question, it should be assumed to be a console input.
import math
a=input("Enter the values seperated by comma")
b=a.split(',')
res=[]
for i in b :
    d=int(i)
    q=math.sqrt((2*50*d)/30)
    a=int(round(q,0))
    res.append(a)
print(res)
    
```

```python
# 15. Write a program to print a right angle triangle with alphabets.(instead of ‘*’, Alphabet).
# Hint: How to do it for
# *****
# ****
# ***
# **
# *
n=int(input("Enter number of rows :"))
ch=65
for i in range(n) :
    k=0
    for j in range(i+1) :
        if(k==26) :
            k=0
        print(chr(ch+k),end=" ")
        k=k+1
    print()
print('--------------------')
ch=65
k=0
for i in range(n,0,-1) :
    for j in range(i) :
        if(k==26) :
            k=0
        print(chr(ch+k),end=" ")
        k=k+1
    print()
```
