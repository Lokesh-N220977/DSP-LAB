```python
#1) In the regex created from r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group 0 cover? Group 1? 
# Group 2? 
import re

p = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')
text = "Call me at 415-555-4242 tomorrow."

match_obj = p.search(text)
print(f"Group 0 (Full): {match_obj.group(0)}")
print(f"Group 1 (Area): {match_obj.group(1)}")
print(f"Group 2 (Local): {match_obj.group(2)}")
# print(f"Group 2 (Local): {match_obj.group(3)}") no group 3 so we will get an error
```

```python
#practice
import re

# We added parentheses inside the second group
# Group 1: (\d\d\d)
# Group 2: ((\d\d\d)-(\d\d\d\d))  <-- The "outer" local group
# Group 3: (\d\d\d)               <-- The "inner" prefix
# Group 4: (\d\d\d\d)             <-- The "inner" line number
p = re.compile(r'(\d\d\d)-((\d\d\d)-(\d\d\d\d))')
text = "Call me at 415-555-4242 tomorrow."

match_obj = p.search(text)

if match_obj:
    print(f"Group 0: {match_obj.group(0)}") # Full match
    print(f"Group 1: {match_obj.group(1)}") # Area Code
    print(f"Group 2: {match_obj.group(2)}") # Local (Full)
    print(f"Group 3: {match_obj.group(3)}") # Local (Prefix only)
    print(f"Group 4: {match_obj.group(4)}") # Local (Line only)
```

```python
# 2)Find website URLs that begin with http:// or https://. 
import re

url_pattern = re.compile(r'https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+')

sample_text = """
Check out my portfolio at https://-www.example.com/project. 
You can also reach the dev site at http://localhost:8080 or 
visit the archived page at https://old-site.org/home.
"""

urls = url_pattern.findall(sample_text)

print(f"Found {len(urls)} URLs:")
for url in urls:
    print(f"- {url}")
```

```python
# 3)Clean up dates in different date formats (such as 3/14/2015, 03-14-2015, and 2015/3/14) by 
# replacing them with dates in a single, standard format. 
import re

date_pattern = re.compile(r"""
    (?P<month>\d{1,2})[/-](?P<day>\d{1,2})[/-](?P<year>\d{4})  # MM-DD-YYYY
    |                                                          
    (?P<year2>\d{4})[/-](?P<month2>\d{1,2})[/-](?P<day2>\d{1,2}) # YYYY-MM-DD
    |
    (?P<day3>\d{1,2})[/-](?P<month3>\d{1,2})[/-](?P<year3>\d{4}) #DD-MM-YYYY
""", re.VERBOSE)

def standard_date(match):
    if match.group('year'):
        y, m, d = match.group('year'), match.group('month'), match.group('day')
    elif match.group('year2'):
        y, m, d = match.group('year2'), match.group('month2'), match.group('day2')
    else:
        y, m, d=match.group('year3'), match.group('month3'), match.group('day3')
    
    return f"{y}-{int(m):02d}-{int(d):02d}"

text = "Dates: 3/14/2015, 03-14-2015, and 2015/3/14. 99/99/9999"

clean_text = date_pattern.sub(standard_date, text)

print(f"Original: {text}")
print(f"Cleaned:  {clean_text}")
```

```python
# 4)How would you write a regex that matches a number with commas for every three digits? It 
# must match the following: 
# • '42'  
# • '1,234'  
# • '6,368,745' 
# but not the following:  
# • '12,34,567' (which has only two digits between the commas)  
# • '1234' (which lacks commas) 
import re

pattern = re.compile(r'^\d{1,3}(,\d{3})*$')

test_cases = ['42', '1,234', '6,368,745', '12,34,567', '1234']

for case in test_cases:
    if pattern.match(case):
        print(f"'{case}': Valid")
    else:
        print(f"'{case}': Invalid")
```

```python
# 5)How would you write a regex that matches the full name of someone whose last name is 
# Nakamoto? You can assume that the first name that comes before it will always be one word 
# that begins with a capital letter.  
# The regex must match the following: 
# • 'Satoshi Nakamoto' 
# • 'Alice Nakamoto'  
# • 'RoboCopNakamoto'  
# but not the following: 
# • 'satoshiNakamoto' (where the first name is not capitalized)  
# • 'Mr. Nakamoto' (where the preceding word has a nonletter character) 
# • 'Nakamoto' (which has no first name) 
# • 'Satoshi nakamoto' (where Nakamoto is not capitalized) 
import re

test_cases = [
    'Satoshi Nakamoto', 
    'Alice Nakamoto', 
    'RoboCopNakamoto', 
    'satoshiNakamoto', 
    'Mr. Nakamoto', 
    'Nakamoto', 
    'Satoshi nakamoto'
]

pattern = r'^[A-Z][a-zA-Z]*\s?Nakamoto$'

print("Regex Results:")
for name in test_cases:
    if re.search(pattern, name):
        print(f"✅ MATCH: {name}")
    else:
        print(f"❌ FAIL:  {name}")
```

```python
# 6)How would you write a regex that matches a sentence where the first word is either Alice, Bob, 
# or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or 
# baseballs; and the sentence ends with a period? This regex should be case-insensitive. It must 
# match the following: 
# • 'Alice eats apples.' 
# • 'Bob pets cats.' 
# • 'Carol throws baseballs.' 
# • 'Alice throws Apples.'  
# • 'BOB EATS CATS. 
# ' but not the following:  
# • 'RoboCop eats apples.'  
# • 'ALICE THROWS FOOTBALLS.' 
# • 'Carol eats 7 cats.' 
import re

test_cases = [
    'Alice eats apples.',
    'Bob pets cats.',
    'Carol throws baseballs.',
    'Alice throws Apples.',
    'BOB EATS CATS.',
    'RoboCop eats apples.',
    'ALICE THROWS FOOTBALLS.',
    'Carol eats 7 cats.'
]

pattern = r'^(Alice|Bob|Carol)\s+(eats|pets|throws)\s+(apples|cats|baseballs)\.$'

print("Results:")
for sentence in test_cases:
    if re.match(pattern, sentence, re.IGNORECASE):
        print(f"✅ MATCH: {sentence}")
    else:
        print(f"❌ FAIL:  {sentence}")
```

```python
# 7)Strong Password Detection Write a function that uses regular expressions to make sure the 
# password string it is passed is strong. A strong password is defined as one that is at least eight 
# characters long, contains both uppercase and lowercase characters, and has at least one digit. 
# You may need to test the string against multiple regex patterns to validate its strength.
import re

def is_strong_password(password):
    # Rule 1: At least 8 characters long
    # Rule 2: At least one uppercase
    # Rule 3: At least one lowercase
    # Rule 4: At least one digit
    
    strong_regex = re.compile(r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8,}$')
    
    
    if strong_regex.match(password):
        return True
    return False

# Testing the function
test_passwords = ["abc12345", "ABC12345", "Abcdefgh", "Abc1234", "Abc12345"]

for pw in test_passwords:
    status = "Strong" if is_strong_password(pw) else "Need to improve"
    print(f"{pw:12} -> {status}")
```
